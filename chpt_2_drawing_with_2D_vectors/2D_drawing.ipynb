```python
import matplotlib.pyplot as plt

import numpy as np
from math import sqrt
```

```python
# scatterplot ords drawing a dino vector !
dv = [(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4),
    (-5,3), (-5,2), (-2,2), (-5,1), (-4,0), (-2,1), (-1,0), (0,-3),
    (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1), (6,4), (3,1)
]
```

```python
# functions for drawing, ploting etc
def scatter_plot( x, y):
    plt.scatter(x, y)
    #plt.show()

def draw_segment( x, y, color=None):
    plt.plot(x, y, color=color)
    #plt.show()
    
def draw_origin(ch):
    plt.plot(0,0, ch, color='k')
    
def draw_axes(color=None):
    plt.axhline(0, color=color)
    plt.axvline(0, color=color)

def draw_grid(b=None):
    plt.grid(b=b)
    
def set_bounds(x, y):
    plt.xlim(x, y)
    plt.ylim(x, y)
    plt.xticks(np.arange(x, y))
    plt.yticks(np.arange(x, y))
    

```

```python
# dino plot 
x = [i[0] for i in dv ]
y = [i[1] for i in dv ]

draw_origin('x')
draw_axes(color='k')
draw_grid(b=True)
set_bounds(-8, 8)
scatter_plot(x, y)
draw_segment(x, y)
```

```python
# Points on the graph of y = x2.
vectors = [(x,x**2) for x in range(-10,11)] 
x = [i[0] for i in vectors]
y = [i[1] for i in vectors]

draw_axes(color='b')
scatter_plot(x,y)
draw_segment(x,y)
```

```python
# Vector addition
def add(v1,v2):
    return (v1[0] + v2[0], v1[1] + v2[1])   
origin = (0,0)
ords = [origin, (4,3), origin, (-1,1), origin, add((4,3), (-1,1))]

x = [i[0] for i in ords]
y = [i[1] for i in ords]

draw_origin('x')
draw_axes(color='k')
scatter_plot(x,y)
draw_grid(b=True)
draw_segment(x,y)
```

```python
# Vector addition dino_vectors or Translations

dv2 = [add((-1.5,-2.5), v) for v in dv]

x = [i[0] for i in dv ]
y = [i[1] for i in dv ]

x1 = [i[0] for i in dv2 ]
y1 = [i[1] for i in dv2 ]

draw_origin('x')
draw_axes(color='k')
draw_grid(b=True)
set_bounds(-8, 8)

scatter_plot(x, y)
scatter_plot(x1, y1)
draw_segment(x, y)
draw_segment(x1, y1)

```

```python
# Vector components and lengths.
# It can be useful to think of vectors as sums of vectors which lie in the x and y directions.
# The vector (4,3) can be re-written as the sum (4,0) + (0,3) and known as the X, Y components

ords = [origin, (4,0), origin, (0,3) , origin, add((4,0), (0,3)), (4,0), (4,3)]

x = [i[0] for i in ords]
y = [i[1] for i in ords]

draw_origin('x')
#draw_axes(color='k')
set_bounds(-1, 6)
scatter_plot(x,y)
draw_grid(b=True)
draw_segment(x,y)

# the length is the count of the number of units or ticks or use pythagorean theorem
def length(v):
    return sqrt(v[0]**2 + v[1]**2)
print(length((4,3))) ## = 5



```

```python
#  Multiplying Vectors by Numbers ..  operation of multiplying a vector by a number is called scalar multiplication
#  Scaling the target by a given factor
def multiply(vector, scalar):
    return (round(scalar * vector[0], 2), round(scalar * vector[1], 2))

print(multiply((1.2, -3.1), 6.5))
```

```python
def subtract(v1, v2):
    return (v1[0] - v2[0], v1[1] - v2[1])

print('subtract (-1, 3), (3, -2) ','equals :', subtract((-1, 3), (3, -2)))

ords = [origin, (-1, 3), origin, subtract((-1, 3), (3, -2)), origin, (3, -2)  ]

x = [ i[0] for i in ords ]
y = [ i[1] for i in ords ]

color = ['b', 'r', 'b']

draw_origin('x')
draw_axes(color='k')
set_bounds(-6, 6)
scatter_plot(x,y)
draw_grid(b=True)


draw_segment(x, y, color=[c for c in color])


```
